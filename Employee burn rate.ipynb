```python
#Treating Null values
# Fill all NaN values with 0 for the entire DataFrame
tob_filled_test = df_train.fillna(0)
# Instead of using inplace=True, reassign the result to the DataFrame columns
df_train['Resource Allocation'] = df_train['Resource Allocation'].fillna(df_train['Resource Allocation'].mean())
df_train['Mental Fatigue Score'] = df_train['Mental Fatigue Score'].fillna(df_train['Mental Fatigue Score'].mean())
df_train['Burn Rate'] = df_train['Burn Rate'].fillna(df_train['Burn Rate'].mean())
```

```python
# Drop the 'Date of Joining' column
df_train.drop('Date of Joining', axis=1, inplace=True)
df_test_employee_ids = df_test["Employee ID"]
df_test.drop('Date of Joining', axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

categorical_cols = df_train.select_dtypes(include=['object']).columns

label_encoder = LabelEncoder()
for col in categorical_cols:
    df_train[col] = label_encoder.fit_transform(df_train[col])
df_train
```

```python
categorical_cols = df_test.select_dtypes(include=['object']).columns

label_encoder = LabelEncoder()
for col in categorical_cols:
    df_test[col] = label_encoder.fit_transform(df_test[col])
df_test
```

```python
Xtrain = df_train.drop('Burn Rate', axis=1)
Ytrain = df_train['Burn Rate']
Xtest = df_test
```

```python
# Scale the selected features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(Xtrain)
X_test_scaled = scaler.transform(Xtest)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score

# Initialize and train the model
model = RandomForestRegressor(random_state=42)
model.fit(Xtrain, Ytrain)

# Evaluate on training data
Ytrain_pred = model.predict(Xtrain)
print("Training R2 Score:", r2_score(Ytrain, Ytrain_pred))
y_val_pred = model.predict(Xtest)
r2 = r2_score(Ytrain, Ytrain_pred)
scaled_r2 = 100 * r2
print(f"Scaled R2 Score on validation data: {scaled_r2:.2f}")
```

```python
# Predict Burn Rate for test data
Ytest_pred = model.predict(Xtest)
```

```python
# Create submission DataFrame
submission1 = pd.DataFrame({
    "Employee ID": df_test_employee_ids,
    "Burn Rate": Ytest_pred
})

# Save to CSV
submission1.to_csv("submission1.csv", index=False)
print("Submission file created: 'submission1.csv'")
```
